```python
import pandas as pd

data = pd.read_csv('titanic_train.csv',
                  index_col='PassengerId')

median_fare = round(data['Fare'].median(), 2)
std_fare = round(data['Fare'].std(), 2)

print(f"Медиана: {median_fare}, Стандартное отклонение: {std_fare}")
```

```python
import pandas as pd

data = pd.read_csv('titanic_train.csv',
                  index_col='PassengerId')

young_survived = data[(data['Age'] < 30) & (data['Survived'] == 1)].shape[0]
young_total = data[data['Age'] < 30].shape[0]
young_rate = round(young_survived / young_total * 100, 1)

old_survived = data[(data['Age'] > 60) & (data['Survived'] == 1)].shape[0]
old_total = data[data['Age'] > 60].shape[0]
old_rate = round(old_survived / old_total * 100, 1)

print(f"Молодые (<30): {young_rate}% выживших")
print(f"Пожилые (>60): {old_rate}% выживших")
```

```python
import pandas as pd

data = pd.read_csv('titanic_train.csv',
                  index_col='PassengerId')

male_survived = data[(data['Sex'] == 'male') & (data['Survived'] == 1)].shape[0]
male_total = data[data['Sex'] == 'male'].shape[0]
male_rate = round(male_survived / male_total * 100, 1)

female_survived = data[(data['Sex'] == 'female') & (data['Survived'] == 1)].shape[0]
female_total = data[data['Sex'] == 'female'].shape[0]
female_rate = round(female_survived / female_total * 100, 1)

print(f"Мужчины: {male_rate}% выживших")
print(f"Женщины: {female_rate}% выживших")
```

```python
import pandas as pd
import re

data = pd.read_csv('titanic_train.csv',
                  index_col='PassengerId')

def get_name(full_name):
    match = re.search(r', (Mr\.|Master\.|Dr\.|Rev\.|Col\.|Major\.|Capt\.) ([A-Za-z]+)', full_name)
    if match:
        return match.group(2)
    return None

male_names = data[data['Sex'] == 'male']['Name'].apply(get_name)
male_names = male_names.dropna()
most_common_name = male_names.value_counts().index[0]
print(f"Самое популярное мужское имя: {most_common_name}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('titanic_train.csv',
                  index_col='PassengerId')



fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Распределение возраста
data[data['Survived'] == 0]['Age'].dropna().hist(alpha=0.5, label='Погибшие', ax=ax1)
data[data['Survived'] == 1]['Age'].dropna().hist(alpha=0.5, label='Выжившие', ax=ax1)
ax1.set_title('Распределение возраста')
ax1.legend()

# Распределение стоимости билетов
data[data['Survived'] == 0]['Fare'].hist(alpha=0.5, label='Погибшие', ax=ax2)
data[data['Survived'] == 1]['Fare'].hist(alpha=0.5, label='Выжившие', ax=ax2)
ax2.set_title('Распределение стоимости билетов')
ax2.legend()

plt.tight_layout()
plt.show()

# Средний возраст
mean_age_survived = data[data['Survived'] == 1]['Age'].mean()
mean_age_died = data[data['Survived'] == 0]['Age'].mean()
print(f"Средний возраст выживших: {mean_age_survived:.1f}")
print(f"Средний возраст погибших: {mean_age_died:.1f}")
```

```python
import pandas as pd

data = pd.read_csv('titanic_train.csv',
                  index_col='PassengerId')

age_by_class_sex = data.groupby(['Pclass', 'Sex'])['Age'].mean().unstack()
print("Средний возраст по классам и полу:")
print(age_by_class_sex)

# Общий средний возраст по классам
age_by_class = data.groupby('Pclass')['Age'].mean()
print("\nСредний возраст по классам:")
print(age_by_class)
```
